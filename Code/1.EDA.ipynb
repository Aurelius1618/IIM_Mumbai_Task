```python
## import libraries
import pandas as pd
import warnings
import folium

warnings.filterwarnings('ignore')
```

```python
pd_df = pd.DataFrame()  ## population-demand dataframe
pd_df[['Lat', 'Lon', 'Demand']] = dd_df[['Latitude', 'Longitude', 'Demand']]
pd_df['Population'] = feats_df['Pnum']
```

```python
# Create a map centered in China
map_china = folium.Map(location=[35, 120], zoom_start=4, tiles='OpenStreetMap')

# Add circles of different sizes based on population
for i, row in pd_df.iterrows():
    folium.CircleMarker(
        location=[row["Lat"], row["Lon"]],
        radius=float(row["Population"])/5000,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(map_china)

# Display the map
map_china
```

```python
# Create a map centered in China
map_china = folium.Map(location=[35, 120], zoom_start=4, tiles='OpenStreetMap')

# Add circles of different sizes based on population
for i, row in pd_df.iterrows():
    folium.CircleMarker(
        location=[row["Lat"], row["Lon"]],
        radius=float(row["Demand"])/5000,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(map_china)

# Display the map
map_china
```
